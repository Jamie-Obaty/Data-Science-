# Matplotlib Package

```python
# load matplotlib.pyplot , pandas, and numpy
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import math
```

```python
# Create array for x- and y- coordinates

x = [1,2,3,4]
y = [5,1,4,2]
# Create 5x6 figure
plt.figure(figsize=[9,6]) # or [6,5]
#Create a line plot
plt.plot(x, y)
```

```python
# Create an equally spaced array with 100 elements form -5 to 5 
x = np.linspace(-5,5, num=100)
# plots the graph for y = x^2
plt.plot(x, x*x)
#plots the graph for y =-4x -4
plt.plot(x, -4*x -  4)
```

```python
# load fatalities.csv data
df2 = pd.read_csv('fatalities.csv')
df2.head()
```

```python
#create scatterplot for total highway fatalities
plt.scatter(df2['Year'], df2['Total'], label= 'Total')
# or plot
#plt.plot(df2['Year'], df2['Total'], label= 'Total')
# Create scatterplot for alcohol-related fatalities
plt.scatter(df2['Year'], df2['Alcohol-related'], label= 'Alcohol-related')
#or plot
#plt.plot(df2['Year'], df2['Alcohol-related'], label= 'Alcohol-related')
# Create title, font size set 20
plt.title("Alcohol-related fatalities on highway", fontsize=20)
#create label for x-axis font size set to 14
plt.xlabel('Year',fontsize=14)
#Create label for y-axis, font size set to 14
plt.ylabel('Highway fatalities',fontsize=14)
#Create annotation for point (1984, 24762) with text at (1986, 300000)
plt.annotate('Drinking age changed to 21', arrowprops =dict(),
             xy=(1984, 24762),
             xytext=(1986, 30000))
#Create legend
plt.legend()
```

```python
df =pd.read_csv('oldfaithfulCluster.csv')
df.head()                
```

```python
#Define groups according to clusters
group1 = df[df['Cluster'] ==1]
group0 = df[df['Cluster'] ==0]

#create Scatterplots
plt.scatter(group1['Eruption'], group1['Waiting'],
            label ='1', edgecolors='White')
plt.scatter(group0['Eruption'], group0['Waiting'],
            label ='0', edgecolors='White')

# Add axes label and title
plt.xlabel('Eruption time', fontsize= 14)
plt.ylabel('Waiting time', fontsize = 14)
plt.title("Eruption time vs waiting time", fontsize=20)

#Add legend
plt.legend(title="Cluster")
```

### Multiple Plot

```python
df = pd.read_csv('oldfaithfulCluster.csv')
```

```python
plt.subplot(2,1,1)
plt.scatter(df['Eruption'], df['Waiting'])
plt.title('Eruption time vs waiting time', fontsize = 20, c='black')
plt.ylabel('Waiting time', fontsize = 14)

plt.subplot(2,1,2)
group1 = df[df['Cluster'] == 1]
group0 = df[df['Cluster'] == 0]
plt.scatter(group1['Eruption'], group1['Waiting'], 
            label = '1', edgecolors='white')
plt.scatter(group0['Eruption'], group0['Waiting'], 
            label = '0', edgecolors='white')
plt.xlabel('Eruption time', fontsize = 14)
plt.ylabel('Waiting time', fontsize = 14)
plt.legend()
```
