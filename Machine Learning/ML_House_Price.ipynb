```python
# Loads necessary packages
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.neural_network import MLPRegressor
import matplotlib_inline.backend_inline
matplotlib_inline.backend_inline.set_matplotlib_formats('svg')
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
```

```python
# Loads the homes.csv dataset
homes = pd.read_csv('homes.csv')
homes
```

```python
# Loads predictor and target variables
X = homes[['Bed','Floor', 'Bath']]
y = homes[['Price']]
```

```python
# Splits the data into training and test sets
X_train, X_test, y_train, y_test = train_test_split(X, np.ravel(y),random_state=123)
```

```python
# Initializes and trains a multilayer perceptron regressor model on the training and validation sets
# This cell takes a long to run.
mlpReg_train = MLPRegressor(random_state=1, max_iter=500000, hidden_layer_sizes=[3,2]).fit(X_train, np.ravel(y_train))
mlpReg_validation = MLPRegressor(random_state=1, max_iter=500000, hidden_layer_sizes=[1]).fit(X_test, np.ravel(y_test))
```

```python
# Predicts the price of a 5 bedroom house with 2,896 sq ft
y_pred = mlpReg_train.predict([[5, 2.896, 4]])
y_pred
```

```python
# Plots the loss curves for the training and validation sets
f, ax = plt.subplots(1, 1)
sns.lineplot(x=range(len(mlpReg_train.loss_curve_)),y = mlpReg_train.loss_curve_, label='Training')
sns.lineplot(x=range(len(mlpReg_validation.loss_curve_)),y = mlpReg_validation.loss_curve_, label='Validation')
ax.set_xlabel('Epochs', fontsize=14)
ax.set_ylabel('Loss', fontsize=14)
```

```python
# Obtains the final weights and biases
print("Weights:", mlpReg_train.coefs_)
print("Bias:", mlpReg_train.intercepts_)
```
